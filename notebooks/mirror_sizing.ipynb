```python
from pygmid import Lookup as lk
from scipy.io import loadmat
import numpy as np
import pandas as pd
```

```python
n = lk('./nfet_03v3.mat')
p = lk('./pfet_03v3.mat')
```

```python
I_ref = 100e-6
```

```python
for L in [0.28, 0.5, 0.7, 1]:
    for gm_id in [6]:
        jd = p.lookup('ID_W', GM_ID=gm_id, L=L)
        w_ref = I_ref/jd
        gm = gm_id * I_ref

        
        print(
            f"L={L:.2f} um, gm/Id={gm_id}",
            f"JD={jd:.3e} A/um",
            f"Wref=Wout={w_ref:.2f} um",
            f"gm={gm*1e3:.2f} mS"
        )
```

```python
#size nmos
for Ln in [0.28, 0.5, 0.7, 1]:
    for gm_id_n in [6]:
        jdn = n.lookup('ID_W', GM_ID=gm_id_n, L=Ln)
        wn_ref = I_ref/jdn
        vgsn = n.look_upVGS(GM_ID=gm_id, L=L, VDS = 1.65, VSB = 0)
        gmn = gm_id_n * I_ref

        
        print(
            f"L={Ln:.2f} um, gm/Id={gm_id_n}",
            f"JD={jdn:.3e} A/um",
            f"Wref=Wout={wn_ref:.2f} um",
            f"VSG={vgsn:.3f} V",
            f"gm={gmn*1e3:.2f} mS"
        )
```
