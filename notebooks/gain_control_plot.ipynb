```python
#Motivation for dropping gain control/programmabiltiy feature
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Width_effective = np.array([36,72,108]) #um
gain_250u_VV = np.array([1.78, 2.15, 2.34])
gain_500u_VV = np.array([2.7,3.5,3.9])
gain_750u_VV = np.array([3.4,4.45,5.1])

gain_250u_db = 20*np.log10(gain_250u_VV)
gain_500u_db = 20*np.log10(gain_500u_VV)
gain_750u_db = 20*np.log10(gain_750u_VV)

tail_currents = np.array([250, 500, 750])
gain_current_VV = np.array([1.78, 2.7, 3.4])
gain_current_db = 20*np.log10(gain_current_VV)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5), sharey=True)

axes[0].plot(Width_effective, gain_250u_db, marker="o", label="250uA tail")
axes[0].plot(Width_effective, gain_750u_db, marker="x", label="750uA tail")
axes[0].plot(Width_effective, gain_500u_db, marker="^", label="500uA tail")

axes[0].set_xlabel("Effective RF width (um)")
axes[0].set_ylabel("Conversion gain (dB)")
axes[0].set_title("Fixed current, variable width")

axes[0].grid(True)
axes[0].legend(loc="best")

axes[1].plot(tail_currents, gain_current_db, marker="o")
axes[1].set_xlabel("Tail Current (uA)")
axes[1].set_ylabel("Conversion gain (dB)")
axes[1].set_title("Fixed width, var. current")
axes[1].grid(True)

plt.show()
fig.tight_layout()

```
